# Text classification - UIT-VSMEC

```python
!mkdir ./datasets/UIT-VSMEC
```

```python
from datasets import load_dataset

ds = load_dataset("Hamana0509/UIT-VSMEC", )

train_dataset = ds['train']
test_dataset = ds['test']

train_dataset = train_dataset.rename_column("Emotion", "Label")
test_dataset = test_dataset.rename_column("Emotion", "Label")

label_mapping = {
    "Other": 0,
    "Disgust": 1,
    "Enjoyment": 2,
    "Anger": 3,
    "Surprise": 4,
    "Sadness": 5,
    "Fear": 6,
}

train_dataset = train_dataset.map(lambda x: {"Label": label_mapping[x["Label"]]})
test_dataset = test_dataset.map(lambda x: {"Label": label_mapping[x["Label"]]})

train_dataset.to_json("./datasets/UIT-VSMEC/UIT-VSMEC_train.json", orient='records', lines=True, force_ascii=False)
test_dataset.to_json("./datasets/UIT-VSMEC/UIT-VSMEC_test.json", orient='records', lines=True, force_ascii=False)
```

```python
!melt --wtype gemini \
              --model_name gemini-2.0-flash-exp \
              --dataset_name UIT-VSMEC \
              --fewshot_prompting True \
              --lang vi \
              --seed 42
```
